```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("comparison_results.csv")

methods = ["multiUserMaximize","heuristic_groups", "heuristic_destination_time", "heuristic_mix"]
num_runs = len(df)
x = np.arange(num_runs)

bar_width = 0.1
offsets = [-1.5, -0.5, 0.5, 1.5, 2.5]  

plt.figure(figsize=(12, 6))


for i, method in enumerate(methods):
    plt.bar(x + offsets[i]*bar_width, df[method], width=bar_width, label=method.replace("_", " ").title())


plt.xlabel("Input numbers", fontsize=12)
plt.ylabel("Total Passengers Assigned", fontsize=12)
plt.title("Comparison of Optimization and Heuristics (Bar Chart)", fontsize=14)
plt.xticks(x, [f"Input {i+1}" for i in x])
plt.legend(loc="lower left", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()


plt.savefig("comparison_bar_chart.png")
plt.show()
```
